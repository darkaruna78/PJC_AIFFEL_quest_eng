```python
'''
Created on 2026. 1. 16.

@author: chaliepark
'''

import time
import pandas as pan

bank_id={"id":[110],"bank_name":["신한"]}
sinhan_account_seq=100000

def create_account_number(bank_name="신한"):
    ##계좌번호:bank_id(3)-branch(2)-account_num(6)
    global sinhan_account_seq
    sinhan_account_seq +=1
    return str(bank_id["id"][bank_id["bank_name"].index("신한")]) + "-" + "01" +  "-" +str(sinhan_account_seq)


class Account():
    _bank_name="신한"
    _account={ ## pk:accnt_num
        "bank_name":[], ##은행이름
        "accnt_num":[], ##계좌번호
        "clnt_name":[], ##예금주
        "total_money":[] ##잔액
    }
    _account_size=0
    _deposit_withdraw_history={ ## pk:accnt_num,dpst_wthdr_type,date_time
        "accnt_num":[],
        "dpst_wthdr_type":[],
        "date_time":[],
        "money":[]
    }
    def __init__(self,clnt_name,money):
        self.account_num=create_account_number()
        self.account_id=len(self._account["accnt_num"])
        self._account["bank_name"].append(self._bank_name)
        self._account["clnt_name"].append(clnt_name)
        self._account["accnt_num"].append(self.account_num)
        self._account["total_money"].append(money)
        if 0<money:
            self.deposit(self.account_num,money)

    def get_account_num(self):
        self._account_num=len(self._account["accnt_num"])
        return self._account_num

    def update_account_money(self,money):
        self._account["total_money"][self.account_id] = self._account["total_money"][self.account_id] + money

    def deposit(self,accnt_num,dpst_money):
        if dpst_money >=1: ## 입금이 1원 보다클 때
            self._deposit_withdraw_history["accnt_num"].append(accnt_num)
            self._deposit_withdraw_history["dpst_wthdr_type"].append("입금")
            self._deposit_withdraw_history["money"].append(dpst_money)
            self._deposit_withdraw_history["date_time"].append(time.strftime('%Y-%m-%d %H:%M:%S', time.localtime()))

            self.update_account_money(dpst_money)
        else:
            print("입금하려는금액이 너무 작습니다")

        dpst_hist_df = pan.DataFrame(self._deposit_withdraw_history)
        if len(dpst_hist_df.query('accnt_num == "{}" and dpst_wthdr_type == "입금"'.format(self.account_num)))% 5 ==0:
            ## 1%이자 입금
            print("이자 입금되었습니다")
            self._account["total_money"][self.account_id] = self._account["total_money"][self.account_id] + (self._account["total_money"][self.account_id]*0.01)

    def withdraw(self,accnt_num,wthdr_money):
        if self._account["total_money"][self._account["accnt_num"].index(accnt_num)]>=wthdr_money:
            self._deposit_withdraw_history["accnt_num"].append(accnt_num)
            self._deposit_withdraw_history["dpst_wthdr_type"].append("출금")
            self._deposit_withdraw_history["money"].append(wthdr_money)
            self._deposit_withdraw_history["date_time"].append(time.strftime('%Y-%m-%d %H:%M:%S', time.localtime()))

            self.update_account_money(wthdr_money*-1)
        else:
            print("출금하려는금액이 너무 큽니다.")
    #def transfer(self):
        # 계좌이체

    def display_all_info(self): ## display_info
        account_df = pan.DataFrame(self._account)
        return account_df

    def display_info(self):
        return f"은행이름:{self._account['bank_name'][self.account_id]}, 예금주:{self._account['clnt_name'][self.account_id]}, 계좌번호:{self._account['accnt_num'][self.account_id]}, 잔고:{self._account['total_money'][self.account_id]:,}원"

    def display_deposit_history(self):
        #print(self._deposit_withdraw_history)
        dpst_hist_df = pan.DataFrame(self._deposit_withdraw_history)
        return dpst_hist_df.query('dpst_wthdr_type == "입금"')
    def display_withdraw_history(self):
        dpst_hist_df = pan.DataFrame(self._deposit_withdraw_history)
        return dpst_hist_df.query('dpst_wthdr_type == "출금"')


b =Account("박진철",3000)
a =Account("로제",1000000)
t =Account("카리나",1000000)

b.get_account_num()
b.display_all_info()


b.get_account_num()
b.deposit(b.account_num,200)
b.deposit(b.account_num,200)
b.deposit(b.account_num,200)
b.deposit(b.account_num,200)
b.deposit(b.account_num,200)
b.deposit(b.account_num,200)
print(b.display_info())

b.display_all_info()
print(b.display_deposit_history())
```
